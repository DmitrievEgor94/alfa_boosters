```python
def truncate(x, num_last_transactions=750):
    return x.values.transpose()[:, :num_last_transactions].tolist()


def transform_transactions_to_sequences(transactions_frame: pd.DataFrame,atrs,
                                        num_last_transactions=750) -> pd.DataFrame:
    return transactions_frame \
        .groupby(['session_id'])[atrs] \
        .apply(lambda x: truncate(x, num_last_transactions=num_last_transactions)) \
        .reset_index().rename(columns={0: 'sequences'})
```

```python
def pad_sequence(array, max_len) -> np.array:
    add_zeros = int(max_len - len(array[0]))
    return np.array([list(x) + [0] * add_zeros for x in array])



def create_padded_buckets(frame_of_sequences: pd.DataFrame, bucket_info: Dict[int, int],
                          save_to_file_path=None, has_target=True):
    frame_of_sequences['bucket_idx'] = frame_of_sequences.sequence_length.map(bucket_info)
        
        
    padded_seq = []
    targets = []
    lengths = []
    sessions = []
    client_pin = []
    timestamp = []

    sum_sessions = 0
    
    from tqdm import tqdm

    for size, bucket in frame_of_sequences.groupby('bucket_idx'):
        padded_sequences = bucket.sequences.apply(lambda x: pad_sequence(x, size)).values
        padded_sequences = np.array([np.array(x) for x in padded_sequences])
        padded_seq.append(padded_sequences)

        if has_target:
            targets.append(bucket.multi_class_target.values)
            
        lengths.append(bucket.sequence_length.values)
        sessions.append(bucket.session_id.values)
        client_pin.append(bucket.client.values)
        
        sum_sessions += bucket.session_id.unique().shape[0]
        timestamp.append(bucket.timestamp.values)
        
    
    frame_of_sequences.drop(columns=['bucket_idx'], inplace=True)

    dict_result = {
        'padded_sequences': np.array(padded_seq),
        'targets': np.array(targets) if targets else [],
        'lengths': np.array(lengths),
        'session_id': np.array(sessions),
        'client': np.array(client_pin),
        'timestamp': np.array(timestamp)
    }

    if save_to_file_path:
        with open(save_to_file_path, 'wb') as f:
            pickle.dump(dict_result, f)
    return dict_result


def create_buckets_from_transactions(df, file_name):   
    import tqdm
    import os
    
    atrs = pickle.load(open('parq_atrs.pickle','rb'))
    
    seq = transform_transactions_to_sequences(df, atrs)
    seq['sequence_length'] = seq.sequences.apply(lambda x: len(x[1]))
    
    df = df.groupby('session_id')[['multi_class_target','client', 'timestamp']].first().reset_index()
            
    seq = seq.merge(df.reset_index(), on='session_id')
    
    mapping_seq_len_to_padded_len = pickle.load(open('buckets_info.pkl', 'rb'))
    
    processed_fragment =  create_padded_buckets(seq, mapping_seq_len_to_padded_len, has_target=True, 
                                                )
    
    pd.DataFrame(processed_fragment).to_pickle(file_name)
    
    return processed_fragment
```

```python
def batches_generator(list_of_paths,transaction_features, batch_size=32, shuffle=False, is_infinite=False,
                      verbose=False, device=None, output_format='torch', is_train=True):
    while True:
        if shuffle:
            np.random.shuffle(list_of_paths)

        for path in list_of_paths:
            if verbose:
                print(f'reading {path}')

            with open(path, 'rb') as f:
                data = pickle.load(f)
            padded_sequences, targets, lengths = data['padded_sequences'], data['targets'], data['lengths']
            
            indices = np.arange(len(targets))

            if shuffle:
                np.random.shuffle(indices)
                
                padded_sequences = padded_sequences[indices]
                targets = targets[indices]
                lengths = lengths[indices]

            for idx in range(len(padded_sequences)):

                bucket = padded_sequences[idx]
                
                if is_train:
                    target = targets[idx]
                    length = lengths[idx]
                
                for jdx in range(0, len(bucket), batch_size):
                    batch_sequences = bucket[jdx: jdx + batch_size]
                    
                                        
                    if is_train:
                        batch_targets = target[jdx: jdx + batch_size]
                        batch_length = length[jdx: jdx + batch_size]
                        
                    batch_sequences = [torch.LongTensor(batch_sequences[:, i]).to(device)
                                       for i in range(len(transaction_features))]
                                    
                    if is_train:
                        yield dict(transactions_features=batch_sequences,
                                   label=torch.LongTensor(batch_targets).to(device),
                                   lengths=torch.LongTensor(batch_length).to(device)
                                  )
                    else:
                        yield dict(transactions_features=batch_sequences,
                                   product=torch.LongTensor(batch_products).to(device),
                                   lengths=torch.LongTensor(data['lengths']).to(device)
                                   )
        if not is_infinite:
            break
```

```python
def train_epoch(model, optimizer, dataset_train, atrs, batch_size=64, shuffle=True,
                print_loss_every_n_batches=100, device=None):
    train_generator = batches_generator(dataset_train, atrs, batch_size=batch_size, shuffle=shuffle,
                                        device=device, is_train=True, output_format='torch')
    
    loss_function = nn.CrossEntropyLoss()

    num_batches = 1
    running_loss = 0.0

    model.train()

    for i, batch in enumerate(train_generator):
        
        _, output = model(batch['transactions_features'], batch['lengths'])
#         print(output.shape)
#         print(batch['label'].shape)
        batch_loss = loss_function(output, batch['label'])
        
        if i % 600 == 0:
            val_generator = batches_generator(dataset_valid, atrs, batch_size=batch_size, shuffle=False,
                                      device=device, is_train=True, output_format='torch')
            metrics = []  
            
            dict_vectrs = pd.read_pickle(dataset_valid[0])
            
            for (seqs, targets, lengths) in zip(dict_vectrs['padded_sequences'], dict_vectrs['targets'], dict_vectrs['lengths']):
                batch_sequences = [torch.LongTensor(seqs[:, i]).to(device)
                                                   for i in range(len(atrs))]

                lengths = torch.LongTensor(lengths).to(device)
                targets = torch.LongTensor(targets).to(device)

                outputs, logits = model(batch_sequences, lengths)

                metrics.append(f1_score(targets.detach().cpu().numpy().flatten(), logits.argmax(axis=-1).detach().cpu().numpy().flatten(), average='macro'))
            
#             for batch_t in val_generator:
                
#                 _, output = model(batch_t['transactions_features'], batch_t['lengths'])
#                 metrics.append(f1_score(batch_t['label'].detach().cpu().numpy().flatten(), output.argmax(axis=-1).detach().cpu().numpy().flatten(), average='macro'))
                
            print(np.mean(metrics))

        batch_loss.backward()
        optimizer.step()
        optimizer.zero_grad()

        running_loss += batch_loss

        if num_batches % print_loss_every_n_batches == 0:
            print(f'Training loss after {num_batches} batches: {running_loss / num_batches}', end='\r')
        
        num_batches += 1
    
    print(f'Training loss after epoch: {running_loss / num_batches}', end='\r')
```

```python
class TransactionsRnn(nn.Module):
    def __init__(self, embedding_projections, transaction_features,rnn_units=128, top_classifier_units=32, classes=11):
        super(TransactionsRnn, self).__init__()
                
        self._transaction_cat_embeddings = nn.ModuleList([self._create_embedding_projection(*embedding_projections[feature]) 
                                                          for feature in transaction_features])
                                
        self._gru = nn.GRU(input_size=sum([embedding_projections[x][1] for x in embedding_projections]),
                             hidden_size=rnn_units, batch_first=True, bidirectional=False)
        
        self._hidden_size = rnn_units
                
        self._top_classifier = nn.Linear(in_features=rnn_units, 
                                         out_features=top_classifier_units)
        
        self._intermediate_activation = nn.ReLU()
        
        self._head = nn.Linear(in_features=top_classifier_units, out_features=classes)
        
        self.transaction_atrs = transaction_features
    
    def forward(self, transactions_cat_features, lengths):             
        embeddings = [embedding(transactions_cat_features[i]) for i, embedding in enumerate(self._transaction_cat_embeddings)]
        
        concated_embeddings = torch.cat(embeddings, dim=-1)
        
        packed_inputs = torch.nn.utils.rnn.pack_padded_sequence(concated_embeddings, lengths, batch_first=True, enforce_sorted=False)
            
        _, final_hiddens = self._gru(packed_inputs)
                
        final_hiddens = final_hiddens.view(concated_embeddings.size(0),-1)
                                           
        classification_hidden = self._top_classifier(final_hiddens)
        activation = self._intermediate_activation(classification_hidden)
        
        logit = self._head(activation)
        
        return [final_hiddens, logit]
    
    @classmethod
    def _create_embedding_projection(cls, cardinality, embed_size, padding_idx=0):
        # +1 категория other ещё
        return nn.Embedding(num_embeddings=cardinality + 1, embedding_dim=embed_size, padding_idx=padding_idx)
```

```python
def transform_parquet_for_net(parquet_name, train_test_df):
    import pickle
    
    torch.cuda.empty_cache()
    
    parquet = pd.read_parquet(parquet_name)

    parquet = pd.merge(parquet, train_test_df[['session_id','multi_class_target']], on=['session_id'], how='left')

    parquet.loc[parquet.device_is_webview.isnull(),'device_is_webview'] = 0

    parquet.fillna('nan', inplace=True)

    print('смержил паркет')

    import pickle
    dict_cols_categ = pickle.load(open('dict_usefull_ctgrs','rb'))

    for col in dict_cols_categ:
        parquet[col] = parquet[col].apply(lambda x: 'other' if x not in dict_cols_categ[col] else x)

    lb_enconders_atrs = pickle.load(open('dict_label_enc_atrs.pickle','rb'))    

    for col in lb_enconders_atrs:
#         print(col)
        le_encoder = lb_enconders_atrs[col]

        le_dict = dict(zip(le_encoder.classes_, le_encoder.transform(le_encoder.classes_)))
        parquet[col] = parquet[col].apply(lambda x: le_dict.get(x, len(le_dict) + 1))

    target_encoder = pickle.load(open('target_encoder.pickle','rb'))

    parquet['multi_class_target'] = target_encoder.transform(parquet['multi_class_target'])

    parquet.sort_values(by='timestamp', inplace=True)
    
    parq_last_name = parquet_name.split('/')[-1]
    
    print(parq_last_name)

    dict_vectrs = create_buckets_from_transactions(parquet, f'/data/edmitrie/alfa_boosters/buckets_first_net/{parq_last_name}.pckl')
    
    model = torch.load('torch_1_model.bin')
    
    device = 'cuda'
    
    model = model.to(device)
    
    els = []

    for (seqs, targets, lengths, sessions, clients,timestamps) in zip(dict_vectrs['padded_sequences'], dict_vectrs['targets'], dict_vectrs['lengths'],\
                                                           dict_vectrs['session_id'], dict_vectrs['client'], dict_vectrs['timestamp']):
        
        batch_sequences = [torch.LongTensor(seqs[:, i]).to(device)
                                           for i in range(len(atrs))]

        lengths = torch.LongTensor(lengths).to(device)

        outputs, logits = model(batch_sequences, lengths)

        els.append(pd.DataFrame(np.concatenate([outputs.cpu().detach().numpy(), np.expand_dims(sessions, axis=1), np.expand_dims(clients, axis=1),\
                                                np.expand_dims(targets, axis=1), np.expand_dims(timestamps, axis=1)], axis=1),\
                                columns=list(range(128)) + ['session_id', 'client_pin','target','timestamp']))
    
    res = pd.concat(els)
    
    res['target'] = target_encoder.inverse_transform(res['target'].astype(int))
    
    print(res['target'].unique())
    
    res.to_pickle(f'/data/edmitrie/alfa_boosters/net_atrs_first_net/{parq_last_name}.pckl')
        
    return res
```
